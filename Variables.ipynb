### Variable memory usage


````python
var1 = 10
````

````python
# Check the memory address of var1 by using the following statement
print(hex(id(var1)))
````

````python
var1 = 100
# Check the memory address of var1 again
````

````python
# Write your code here
print(hex(id(var1)))
````

You should see two distinct addresses for var1. Explain why there are two different addresses and what happened to the first one.

There are two different addresses because a new integer object is being created at a different address and reassigning var1 to reference the new address.

````python
var2 = 100
# Write your print statement
print(var2)
# Check the memory address of var2. Did the Python interpreter assign a new memory address or reuse the existing one?
````

````python
# Write your code here
print(hex(id(var2)))
print(hex(id(var1)))
````

The address for var2 and var1 are the same. This is because they are referencing the same integer object as they are the same number with the intent to optimize the memory useage.
